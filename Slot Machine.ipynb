-All code for program is contained in a run_slot_machine() function in the cell below. Running the cell below that actually executes the function.

-While the player can only make integer bets, they can win amounts with up to two decimal places (representing cents) which will be reflected also by their purse total. 

-Keeps running until the player doesn't have enough money for the minimum bet (purse < 2) or 'N' is entered as a bet. 

```python
def run_slot_machine():
    """executes all the code used by the slot machine"""  
    from random import choice
    from decimal import Decimal #used later to prevent numbers with > two decimal places appearing

    faces = ['ðŸŒ', 'ðŸ’', 'ðŸŠ']#creating an array of fruit symbols

    class purse(object):
        cash = 100 #the money the player begins with

        def debit(self, x): 
            """Adds cash to the purse"""
            self.cash -= x

        def credit(self, x):
            """Removes cash from purse"""
            self.cash += x

        def get_balance(self):
            """Checks the balance of the purse"""
            print("You now have:", self.cash)

    purse = purse #creating a purse class object called purse

    class column(object):

        def change_face():
            """Returns a random fruit from faces"""
            return choice(faces)

        def __init__(self, face):
            self.face = face


    class slot(object):
        a = column.change_face() #each slot object begins with three columns with random faces
        column1 = column(a)
        b = column.change_face()
        column2 = column(b)
        c = column.change_face()
        column3 = column(c)
        score = 0 #initialising key variables
        bet = 0

        def show_slot(self):
            """Reveals the faces on the three columns of slot"""
            print(self.column1.face, self.column2.face, self.column3.face)

        def take_bet(self):
            """Takes the players bet"""
            q = True #variable to have loop of asking for bet until a valid bet is entered
            while q == True:
                x = input("How much would you like to bet? ")
                if x == "N":
                    print("")
                    print("You walk away with: ", purse.cash)
                    print("")
                    print("Thanks for playing!")
                    print("====================================")
                    self.bet = 'n'
                    q = False
                else:
                    try:
                        x = int(x) 
                    except ValueError as nonint:
                        print("")
                        print("Please only enter an integer value for bet, or type 'N' to exit.")
                    else:
                        if x > purse.cash:
                            print("")
                            print("You cannot bet more money than you have.")
                        elif x < 2:
                            print("")
                            print("You cannot bet less than 2.")
                        else:
                            self.bet = x
                            q = False #to exit loop as a valid bet has been entered

        def score_slot(self):
            """Calculates the players score for that round based on the faces"""
            if self.column1.face == self.column2.face == self.column3.face: #all matching
                slot.score += 1
            elif self.column1.face == self.column2.face != self.column3.face: #two matching
                self.score += 0
            elif self.column2.face == self.column3.face != self.column1.face: #two matching
                self.score += 0
            elif self.column3.face == self.column1.face != self.column2.face: #two matching
                self.score += 0
            else: #where none are matching
                self.score -= 1


        def pull_lever(self):
            """Randomly changes then shows faces and takes/gives cash based on score"""
            self.column1.face = column.change_face() #randomly changing the faces of the three columns
            self.column2.face = column.change_face()
            self.column3.face = column.change_face()
            self.score = 0
            slot.show_slot(self)
            slot.score_slot(self)
            if self.score == 1: #all matching
                print("Full house!")
                purse.credit(purse, self.bet)
                print("You win:", self.bet)
                print("")
                print("You now have:", purse.cash)
            elif self.score == 0: #two matching
                print("Half house!")
                x = Decimal(self.bet/2) #Rounding off the amount won, x, to the two decimal place
                x = round(x, 2)         #Reference: gist.github.com/jackiekazil/6201722
                purse.credit(purse, x)
                print("You win:", (x))
                print("")
                print("You now have:", purse.cash)
            else: #none matching
                print("Empty house!")
                purse.debit(purse, self.bet)
                print("You lose:", self.bet)
                print("")
                print("You now have:", purse.cash)

    machine = slot #creating a class slot object called machine
    print("==========  SLOT MACHINE  ==========")
    print("")
    print("Minimum bet is 2. Enter 'N' as your bet to exit.")
    print("")
    print("You have:", purse.cash)
    print("")
    while purse.cash > 1: #all below to loop until the player does not have enough money for a new bet
        slot.take_bet(machine)
        print("")
        if machine.bet == 'n':
            break
        slot.pull_lever(machine)
        if purse.cash <= 1:
            print("You are broke. Thank you for playing!") #message on hitting 1 or less cash
            print("====================================")
```
